## 16.8
Connected to the books.db database using SQLAlchemy and printed book titles in alphabetical order.


```python


from sqlalchemy import create_engine, text

engine = create_engine('sqlite:///books.db')

with engine.connect() as conn:
    result = conn.execute(text("SELECT title FROM book ORDER BY title"))
    for row in result:
        print(row[0])
```

```python
import sqlite3
conn = sqlite3.connect('books.db')
cursor = conn.cursor()

cursor.execute('CREATE TABLE IF NOT EXISTS book (title TEXT)')

cursor.execute("INSERT INTO book VALUES ('The Hobbit')")
cursor.execute("INSERT INTO book VALUES ('1984')")
cursor.execute("INSERT INTO book VALUES ('Pride and Prejudice')")

conn.commit()
conn.close()
```
